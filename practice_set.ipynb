Write a Python function to remove a given word from a list and strip it at the same time.


```python
def remove_and_strip(word_list, word):
    new_list = []
    for item in word_list:
        if item.strip() != word:
            new_list.append(item.strip())
    return new_list

words = [" apple ", " banana ", " mango ", " apple "]
result = remove_and_strip(words, "apple")

print(result)
```

```python
def remove(word_list, word):
    return [w.strip() for w in word_list if w.strip() != word]
x=[" java", "python", " c++ ", " python "]
print(remove(x, "python"))
```
